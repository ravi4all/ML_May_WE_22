```python
# Plotly - Data Visualization Library
```

```python
# Plotly works in 2 ways
#  - Plotly Graph Objects - lower level API
#  - Plotly Express - high level API
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import pandas as pd
```

```python
df = pd.read_csv('SaratogaHouses.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
figure = px.bar(df, x="bedrooms", y="fuel")
figure.show()
```

```python
fig = go.Figure()
for c, group in df.groupby('bedrooms'):
    fig.add_trace(go.Bar(x=group['bedrooms'], y=group['fuel']))
fig.show()
```

```python
px.histogram(df, x='price').show()
```

```python
px.bar(df, x='rooms', y='price').show()
```

```python
px.bar(df, x='price', y='rooms', orientation='h').show()
```

```python
go.Figure(
    data = [go.Pie(labels=df['rooms'], values=df['price'])]
).show()
```

```python
px.scatter(df, x='livingArea', y='price', text='rooms').show()
```

```python
px.scatter(df, x='livingArea', y='price', size='rooms', color='fuel', text='rooms').show()
```

```python
px.box(df, x='price').show()
```
