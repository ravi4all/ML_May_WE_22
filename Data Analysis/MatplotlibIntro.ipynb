```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,20)
y = x ** 3
```

```python
# Line Plot
plt.plot(x, y, 'ro--')
```

```python
# Line Plot
# plt.scatter(x, y)
plt.scatter(x, y, s=100, c='red', marker='*')
```

```python
countries = ['ind','us','china','russia','france']
medals = [100, 1500, 2100, 510, 119]
```

```python
plt.bar(countries, medals)
```

```python
plt.figure(figsize=(10,10))
plt.pie(medals, labels=countries, autopct='%1.2f%%', shadow=True, counterclock=False, startangle=90,
       explode=(0,0,0,0.4,0))
plt.legend()
plt.title("Medals Won by Each Country")
plt.show()
```
