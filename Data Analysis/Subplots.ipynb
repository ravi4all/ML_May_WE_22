```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y1 = np.array([3,5,8,10,9,6,12,10,11,20])
y2 = np.array([3,5,8,6,19,6,10,7,8,10])
```

```python
plt.scatter(x,y1)
plt.show()
```

```python
plt.scatter(x,y2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))

ax1.scatter(x, y1, color='green')
ax2.scatter(x, y2, color='red')

ax1.set_title("Plot_1")
ax2.set_title("Plot_2")

plt.show()
```
