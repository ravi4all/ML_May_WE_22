```python
# Blurring of images
```

```python
# Gaussian Blur
# Median Blur
# Bilateral Filter
```

```python
img_blur = cv2.GaussianBlur(img, (7,7), 20)
```

```python
cv2.imshow('output',img_blur)
cv2.waitKey(0)
```

```python
blur_1 = cv2.blur(img, (5,5))
blur_2 = cv2.blur(img, (9,9))
blur_3 = cv2.blur(img, (11,11))
blur_4 = cv2.blur(img, (25,25))
titles = ['5x5', '9x9', '11x11', '25x25']

images = [blur_1, blur_2, blur_3, blur_4]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
blur_1 = cv2.GaussianBlur(img, (5,5), 0)
blur_2 = cv2.GaussianBlur(img, (9,9), 0)
blur_3 = cv2.GaussianBlur(img, (11,11), 0)
blur_4 = cv2.GaussianBlur(img, (25,25), 0)
titles = ['5x5', '9x9', '11x11', '25x25']

images = [blur_1, blur_2, blur_3, blur_4]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
blur_1 = cv2.medianBlur(img, 5)
blur_2 = cv2.medianBlur(img, 9)
blur_3 = cv2.medianBlur(img, 11)
blur_4 = cv2.medianBlur(img, 25)
titles = ['5x5', '9x9', '11x11', '25x25']

images = [blur_1, blur_2, blur_3, blur_4]
for i in range(4):
    cv2.imshow(titles[i], images[i])
    cv2.waitKey(0)
```
